```python
cad1 = input("Inserta la secuencia 1: ").upper()
cad2 = input("Inserta la secuencia 2: ").upper()

n1 = len(cad1)
n2 = len(cad2)
total_cambios = 0  # La inicializamos aquí para evitar errores de ejecución

if n1 != n2:
    print(f"Las cadenas son de diferente longitud por {abs(n1-n2)} caracteres.")
    print(f"La cadena 1 mide {n1} y la cadena 2 mide {n2} nucleótidos.")
else:
    # Verificamos primero si hay alguna diferencia antes de imprimir el encabezado
    # O simplemente imprimimos el encabezado si vas a listar los cambios:
    
    # Creamos una lista para guardar las posiciones y saber si hubo cambios
    posiciones = []
    
    for i in range(n1):
        if cad1[i] != cad2[i]:
            posiciones.append(i)
            total_cambios += 1

    if total_cambios > 0:
        print("---")
        print("Se detectaron diferencias en las secuencias en las posiciones:")
        # Volvemos a iterar o imprimimos los resultados guardados
        for i in posiciones:
            print(f"Posición {i}: {cad1[i]} -> {cad2[i]}")
        print("---")
        print(f"El total de diferencias entre las cadenas es de: {total_cambios}")
    else:
        print("Las cadenas insertadas son similares (idénticas).")
```
